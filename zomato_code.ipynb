```python
test.time_stamp_utc = test.time_stamp_utc.map(lambda x:timestripper(x))

test.time_stamp_utc = pd.to_datetime(test.time_stamp_utc)
test['hour'] = test.time_stamp_utc.dt.hour
```

```python
training = stockout.drop(labels = ['geography', 'time_stamp_utc', 'dt','res_id', 'stockout'], axis = 1)
testing = test.drop(labels = ['geography', 'time_stamp_utc', 'dt', 'res_id'], axis = 1)
target = stockout['stockout']
```

```python
submission = pd.read_csv('submission_online_testcase.csv')
```

```python
submission.tail()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_data, x_test, y_data, y_test = train_test_split(training, target, test_size = 0.1, stratify = target, random_state = 1)
x_data = x_data.reset_index(drop = True)
x_test = x_test.reset_index(drop = True)
y_data = y_data.reset_index(drop = True)
y_test = y_test.reset_index(drop = True)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size = 0.25, stratify = y_data, random_state = 1)
x_train = x_train.reset_index(drop = True)
x_val = x_val.reset_index(drop = True)
y_train = y_train.reset_index(drop = True)
y_val = y_val.reset_index(drop = True)
```

```python
import xgboost as xgb
from sklearn.metrics import f1_score
```

```python
params = {"objective":"multi:softmax",
          "booster":"gbtree",
          "eta":0.1,
          "max_depth": 4,
         "subsample":0.95,
         "silent":0,
         "seed":1,
         "lambda":0.1,
         "alpha":0.1,
         "num_class":2,
         "eval_metric":"accuracy"}

def evalerror(preds, dtrain):
    labels = dtrain.get_label()
    return 'F1-score', f1_score(labels, preds, average = 'macro')

num_boost_round = 2500

dtrain = xgb.DMatrix(x_train, y_train)
dvalid = xgb.DMatrix(x_val, y_val)

watchlist = [(dtrain, 'train'), (dvalid, 'eval')]

booster = xgb.train(params, dtrain, num_boost_round, evals = watchlist, early_stopping_rounds = 10, verbose_eval = 10)
```

```python
performance = booster.predict(xgb.DMatrix(x_test))
```

```python
for i in range(len(performance)):
    if performance[i]>= 0.5:
        performance[i] = 1
    else:
        performance[i] = 0
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print('Accuracy = ', accuracy_score(y_test, performance))
```

```python
print(confusion_matrix(y_test, performance))
```

```python
print(classification_report(y_test, performance))
```

```python
prediction = booster.predict(xgb.DMatrix(testing))

for i in range(len(prediction)):
    if prediction[i] >=0.3:
        prediction[i] = 1
    else:
        prediction[i] = 0
```

```python
submission['stockout'] = pd.DataFrame({'stockout':prediction})
```
